### Exemplo 1: raiz de equação


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def exemplo1_fun(x):
    return x**3 - x - 2

x = np.linspace(-2, 2, 100)
y = exemplo1_fun(x)

plt.plot(x, y)
plt.axhline(0, color='red', linestyle='--')
plt.title('Gráfico da função exemplo1_fun')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

```python
x_0 = 1.521
f_x0 = exemplo1_fun(x_0)
print(f"f({x_0}) = {f_x0}")

plt.plot(x, y)
plt.axhline(0, color='red', linestyle='--')
plt.title('Gráfico da função exemplo1_fun')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.plot(x_0, f_x0, 'ro', label=f'Ponto de teste: ({x_0}, {f_x0:.2f})')
plt.legend()
plt.show()
```

```python
0.1 + 0.2 == 0.3
```

```python
# Veja como são representados os valores anteriores
print("0.1 = {0:.17f}".format(0.1))  # formatação que lembra o printf de c
print("0.2 = {0:.17f}".format(0.2))
print("0.1 + 0.2 = {0:.17f}".format(0.1 + 0.2))
print("0.3 = {0:.17f}".format(0.3))
```

```python
def exemplo2_fun(x):
    return 1 + x + x**2/2 

e_aprox = exemplo2_fun(1)
print(f"Valor aproximado: {e_aprox:.17f}")
print(f"Valor real de e: {np.exp(1):.17f}")
print(f"Erro absoluto: {abs(e_aprox - np.exp(1)):.17f}")
print(f"Erro relativo: {abs(e_aprox - np.exp(1)) / abs(np.exp(1)):.17f}")
```

```python
x = 1e6
f1 = np.sqrt(x**2 + 1) - x
f2 = 1/(np.sqrt(x**2 + 1) + x)

print(f"{f1:.17f}")
print(f"{f2:.17f}")
```
