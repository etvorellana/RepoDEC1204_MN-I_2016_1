## Pseudocódigo — Spline Cúbico Natural

(Adaptado de Burden & Faires)

1. h[i] = x[i+1] - x[i]

2. α[i] = (3/h[i])(y[i+1]-y[i]) - (3/h[i-1])(y[i]-y[i-1])

3. Inicialização:
   l[0]=1, μ[0]=0, z[0]=0

4. Para i=1 até n-1:
   l[i]=2(x[i+1]-x[i-1]) - h[i-1]*μ[i-1]
   μ[i]=h[i]/l[i]
   z[i]=(α[i]-h[i-1]*z[i-1])/l[i]

5. Condições finais:
   l[n]=1, z[n]=0, c[n]=0

6. Para j=n-1 até 0:
   c[j]=z[j]-μ[j]*c[j+1]
   b[j]=(y[j+1]-y[j])/h[j] - h[j]*(c[j+1]+2*c[j])/3
   d[j]=(c[j+1]-c[j])/(3*h[j])
